<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Set data set characteristics.
</span>


```python
img_width = 600
img_height = 450
batch_size = 32 
epochs = 20 
channels = 3
img_shape = (img_width, img_height, channels)
```

<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Acquire images from directory.
</span>


```python
def create_df(base_dir):
    dataset = []

    for class_name in os.listdir(base_dir):
        class_path = os.path.join(base_dir, class_name)

        if os.path.isdir(class_path):
            for img_name in os.listdir(class_path):
                if img_name.endswith('.jpg') or img_name.endswith('.jpeg') or img_name.endswith('.png'):
                    img_path = os.path.join(class_path, img_name)

                    img_data = Image.open(img_path).convert('RGB')
                    img_data = img_data.resize((img_width, img_height))
                    img_array = np.array(img_data)

                    dataset.append([img_name, img_array, class_name])

    return pd.DataFrame(dataset, columns=['img_name', 'img_array', 'disease'])

base_dir_train = '../dataset/Train'
base_dir_test = '../dataset/Test'

df_train = create_df(base_dir_train)
df_test = create_df(base_dir_test)
```

```python
classes = os.listdir(base_dir_train)
plt.figure(figsize=(15, 10))

for i, class_name in enumerate(classes):
    class_path = os.path.join(base_dir, class_name)
    
    if os.path.isdir(class_path):
        img_list = os.listdir(class_path)
        
        rand_img = random.choice(img_list)
        img_path = os.path.join(class_path, rand_img)
        img = Image.open(img_path)
                
        plt.subplot(2, len(classes) // 2 + len(classes) % 2, i + 1)
        plt.imshow(img)
        plt.title(class_name)
        plt.axis('off') 

plt.tight_layout()
plt.show()   
```

    • Lesion distribution.
</span>

```python
def class_distribution(df, dataset_text):
    plt.figure(figsize=(15,8))
    ax = sns.countplot(x=df.disease)
    ax.set_xlabel("Diseases")
    plt.title(f"The Number Of Samples per Class in {dataset_text} set")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
class_distribution(df_train, dataset_text='Training')
class_distribution(df_test, dataset_text='Test')
```
