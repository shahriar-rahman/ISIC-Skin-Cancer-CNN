# <div style="text-align:left; border-radius:10px 50px; padding:15px; color:white; margin:0; font-size:95%; font-family:Pacifico; background-color:indigo; overflow:hidden"><b> • Assemble Dataframe </b></div>


<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Set data set characteristics.
</span>


```python
img_width = 150
img_height = 150
batch_size = 32 
epochs = 18
channels = 3
num_classes = 9
input_shape = (img_width, img_height, channels)
```

<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Acquire images from directory.
</span>


```python
def create_df(base_dir):
    dataset = []

    for class_name in os.listdir(base_dir):
        class_path = os.path.join(base_dir, class_name)

        if os.path.isdir(class_path):
            for img_name in os.listdir(class_path):
                if img_name.endswith('.jpg') or img_name.endswith('.jpeg') or img_name.endswith('.png'):
                    img_path = os.path.join(class_path, img_name)

                    img_data = Image.open(img_path).convert('RGB')
                    img_data = img_data.resize((img_width, img_height))
                    img_array = np.array(img_data)

                    dataset.append([img_name, img_array, class_name])

    return pd.DataFrame(dataset, columns=['img_name', 'img_array', 'disease'])

base_dir_train = '../dataset/Train'
base_dir_test = '../dataset/Test'

df_train = create_df(base_dir_train)
df_test = create_df(base_dir_test)
```

# <div style="text-align:left; border-radius:10px 50px; padding:15px; color:white; margin:0; font-size:95%; font-family:Pacifico; background-color:indigo; overflow:hidden"><b> • Feature Transformation </b></div>


<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Organize data.
</span>

```python
images_train = np.array(df_train['img_array'].tolist())  #nP
labels_train = df_train['disease'].values  
```

<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Allocation and Transformation.
</span>

```python
# Encode the labels to int
label_encoder = {label: i for i, label in enumerate(np.unique(labels_train))}
encoded_labels = np.array([label_encoder[label] for label in labels_train])

# OHE the labels (multi-class)
encoded_labels = tf.keras.utils.to_categorical(encoded_labels)

# Split
X_train, X_val, y_train, y_val = train_test_split(images_train, encoded_labels, test_size=0.1, random_state=42)

# Check 
print(f"Training data shape: {X_train.shape}, Labels shape: {y_train.shape}")
print(f"Validation data shape: {X_val.shape}, Labels shape: {y_val.shape}")
```

<a id="A"></a>
# <div style="text-align:left; border-radius:10px 50px; padding:15px; color:white; margin:0; font-size:95%; font-family:Pacifico; background-color:indigo; overflow:hidden"><b> • CNN Architecture </b></div>


<span style="font-family:Cambria; color:#0261ab; font-size: 20px; text-shadow: 0.5px 0.5px black"> 
    • Construct the architecture.
</span>

```python
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Conv2D, BatchNormalization, MaxPool2D, Dropout, Flatten, Dense, AveragePooling2D

def construct_cnn(input_shape):
    model = models.Sequential()     
    
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, strides=(1, 1), padding="same")),
    model.add(AveragePooling2D((2, 2), strides=None, padding="valid")),
    
    model.add(Conv2D(64, (3, 3), activation='relu', strides=(1, 1), padding="same")),
    model.add(AveragePooling2D((2, 2), strides=None, padding="valid")),

    model.add(Conv2D(128, (3, 3), activation='relu', strides=(1, 1), padding="valid")),
    model.add(AveragePooling2D((2, 2), strides=None, padding="valid")),
    
    model.add(Flatten()), 
    model.add(Dense(128, activation='relu')),
    model.add(Dense(len(label_encoder), activation='softmax'))
    
    return model
```

```python
model = construct_cnn(input_shape)
    
model.summary()
```

<a id="A"></a>
# <div style="text-align:left; border-radius:10px 50px; padding:15px; color:white; margin:0; font-size:95%; font-family:Pacifico; background-color:indigo; overflow:hidden"><b> • Model Training </b></div>


```python
from tensorflow.keras.optimizers import Adam 
from keras.callbacks import EarlyStopping

model.compile(optimizer='rmsprop', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

callback = keras.callbacks.EarlyStopping(monitor='loss', patience=5)
```

```python

history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val), verbose=2, 
                    callbacks=[callback])
                    
```

<a id="A"></a>
# <div style="text-align:left; border-radius:10px 50px; padding:15px; color:white; margin:0; font-size:95%; font-family:Pacifico; background-color:indigo; overflow:hidden"><b> • Evaluations </b></div>


```python
# Evaluate on the validation sets
val_loss, val_acc = model.evaluate(X_val, y_val)
print(f"Validation accuracy: {val_acc * 100:.2f}%")
```
